```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
%matplotlib inline
```

```python
UK=df[df['Chapter'].str.contains("UK")==True]
MY=df[df['Chapter'].str.contains("MY")==True]
HK=df[df['Chapter'].str.contains("HK")==True]
NA=df[df['Chapter'].str.contains("NorthAm")==True]
SG=df[df['Chapter'].str.contains("SG")==True]
```

```python
A=list(NA.groupby(['College/University']).size().sort_values(ascending=True))
A
```

```python
A=list(NA.groupby(['College/University']).size().sort_values(ascending=True))
#B=NA.groupby(['College/University']).size().sort_values(ascending=True)
S = ['%.2f' % (i/sum(A)*100) for i in A]
C=[(i,j) for i, j in zip(A, S)]

plt.style.use('seaborn')
NAUni=NA.groupby(['College/University']).size().sort_values(ascending=True).plot(kind='barh',figsize=(10,10),title='North America Universities')

for y, (f,g) in enumerate(C):
    plt.annotate((str(f)+ ', ' + str(g) +'%'), xy=(f, y),va='center')
```
