```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.backends.backend_pdf import PdfPages
%matplotlib inline
```

```python
df = pd.read_excel('I22016.xlsx')
```

```python
df.head()
```

```python
df.isnull().values.any()
```

```python
df.dropna(inplace=True)
```

```python
df.tail()
```

```python
df.groupby(['Gender']).size()
```

```python
plt.style.use('fivethirtyeight')
df.groupby(['Gender']).size().plot(kind='pie',figsize=(8,11),title='# of Males and Females',autopct='%1.3f%%')
plt.axes().set_ylabel('')
plt.axis("equal")
plt.show()
```

```python
len(df['University/College'].unique())
```

```python
df['University/College'].unique()
```

```python
plt.style.use('seaborn-notebook')
df.groupby(['University/College','Gender'])['Gender'].size().sort_values(ascending=False).unstack('Gender').fillna(0).plot(kind='barh',stacked=True,figsize=(12,15))
#df1[['Male','Female']].plot(kind='barh',stacked=True,figsize=(12,15))
plt.xlabel('# of people attended')
plt.grid(False)
plt.show()
```

## Breaks down male-to-female ratio in the charts below

```python
df2=df.groupby(['University/College','Gender'])['Gender'].size().unstack('Gender',fill_value=0).sort_values(by=['Male','Female'],ascending=True)
df2.plot(kind='barh',stacked=True,figsize=(12,15))
plt.xlabel('# of people attended')
plt.grid(False)
plt.show()
```
